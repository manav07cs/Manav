```python
# 21. Convert Launch_Date to datetime
df['Launch_Date'] = pd.to_datetime(df['Launch_Date'], errors='coerce')
df['Launch_Date']
```

```python
# 22. Remove duplicate records
df = df.drop_duplicates()
df
```

```python
# 23. Standardize categorical values
df['Engine_Type'] = df['Engine_Types'].str.title()
df['Fuel_Type'] = df['Fuel_Types'].str.title()
df['Engine_Types'].str.title()
df['Fuel_Types'].str.title()
```

```python
# 24. Treat unrealistic values
df = df[(df['passengers'] > 0) & (df['crew'] > 0)]
(df['passengers'] > 0) & (df['crew'] > 0)
```

```python
# 25. Rename columns
df.rename(columns={
    'Cruise_Line': 'Cruise_Lines',
    'Engine_Type': 'Engine_Types',
    'Fuel_Type': 'Fuel_Types',
    'Hull_Material': 'Hull_Materials',
    'Interior_Style': 'Interior_Styles'
}, inplace=True)
df
```

```python
# 26. Ship Age Category
df['Ship_Age_Category'] = pd.cut(
    df['Age'],
    bins=[0,10,20,100],
    labels=['New','Mid-Age','Old']
)
df['Ship_Age_Category']
```

```python
# 27. Passenger Load
df['Passenger_Load'] = pd.cut(
    df['passengers'],
    bins=[0,1500,2500,10000],
    labels=['Low','Medium','High']
)
df['Passenger_Load']
```

```python
# 28. Crew to Passenger Ratio
df['Crew_to_Passenger_Ratio'] = df['crew'] / df['passengers']
df['Crew_to_Passenger_Ratio']
```

```python
# 29. Length Category
df['Length_Category'] = pd.cut(
    df['length'],
    bins=[0,800,900,2000],
    labels=['Small','Medium','Large']
)
df['Length_Category']
```

```python
# 30. Average passengers by cruise line
df.groupby('Cruise_line')['passengers'].mean()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
# 31. Average crew size by engine type
df.groupby('Engine_Types')['crew'].mean().round(2)
```

```python
# 32. Average passenger density by fuel type
df.groupby('Fuel_Types')['passenger_density'].mean()
```

```python
# 33. Average ship length by hull material
df.groupby('Hull_Materials')['length'].mean()
```

```python
# 34. Average cabins per cruise line
df.groupby('Cruise_line')['cabins'].mean()
```

```python
# 35. Electric vs Diesel passenger capacity
df.groupby('Fuel_Types')['passengers'].mean()
```

```python
# 36. Ship age vs passenger density
df.groupby('Ship_Age_Category', observed=False)['passenger_density'].mean()
```

```python
# 37. Luxury vs Modern interior on length
df.groupby('Interior_Styles')['length'].mean()
```

```python
# 38. Crew size variation across cruise lines
df.groupby('Cruise_line')['crew'].describe()
```

```python
# Ensure df is a proper copy (recommended after filtering)
df = df.copy()
# 39. Ships launched before vs after 2000
df.loc[:, 'Launch_Year_Group'] = np.where(
    df['Launch_Date'].dt.year < 2000,
    'Before 2000',
    'After 2000'
)
df.groupby('Launch_Year_Group')[['passengers', 'crew', 'length']].mean()
```
